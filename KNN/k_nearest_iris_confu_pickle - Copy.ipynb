```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=iris.data
y=iris.target
```

```python
df=pd.DataFrame(x)
df
```

```python
labels=['sepal length (cm)',
 'sepal width (cm)',
 'petal length (cm)',
 'petal width (cm)']
df=pd.DataFrame(x,columns=labels)
df
```

```python
df['target']=y
df
```

```python
df
```

```python
df1=df[df.target==0]
df1
```

```python
df2=df[df.target==1]
```

```python
df3=df[df.target==2]
```

```python
mylabels1=['setosa']
mylabels2=['versicolor']
mylabels3=['virginicia']
```

```python

xpoints1 = df1['sepal length (cm)']
ypoints1 = df1['sepal width (cm)']

xpoints2 = df2['sepal length (cm)']
ypoints2 = df2['sepal width (cm)']

xpoints3 = df3['sepal length (cm)']
ypoints3 = df3['sepal width (cm)']

plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.title("K_N_earest")

plt.scatter(xpoints1,ypoints1,label=mylabels1)
plt.scatter(xpoints2,ypoints2,label=mylabels2)
plt.scatter(xpoints3,ypoints3,label=mylabels3)

plt.legend()
plt.show()

```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=10)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_test[:10]
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred[:10]
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(5,4))
sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import pickle
with open('model.pickle','wb') as file:
    pickle.dump(model,file)
```

```python
with open('model.pickle2','rb') as file:
    classifier=pickle.load(file)
```

```python
y_pred = classifier.predict(x_test)   # use classifier as a model and predicted
```

```python
y_pred  
```

```python
classifier.score(x_test,y_test)  #find accuracy
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(cm,annot=True)
plt.xlabel('Predict')
plt.ylabel('Actl')
plt.show()
```
