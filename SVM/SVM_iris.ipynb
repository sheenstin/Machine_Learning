```python
# set x and y
x = iris.data
y = iris.target
```

```python
df=pd.DataFrame(x,columns=iris.feature_names)
df
```

```python
df['target']=y
df
```

```python
df1=df[df.target==0]
df1
```

```python
df2=df[df.target==1]
```

```python
df3=df[df.target==2]
```

```python
iris.feature_names
```

```python
m_label1=['setosa']
m_label2=['versicolor']
m_label3=['virginica']
```

```python
x1points = df1['petal length (cm)']
y1points = df1['petal width (cm)']

x2points = df2['petal length (cm)']
y2points = df2['petal width (cm)']

x3points = df3['petal length (cm)']
y3points = df3['petal width (cm)']

plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.title("SVM_iris")

plt.scatter(x1points,y1points,label=m_label1)
plt.scatter(x2points,y2points,label=m_label2)
plt.scatter(x3points,y3points,label=m_label3)

plt.legend()
plt.show()
```

```python
x1points = df1['sepal length (cm)']
y1points = df1['sepal width (cm)']

x2points = df2['sepal length (cm)']
y2points = df2['sepal width (cm)']

x3points = df3['sepal length (cm)']
y3points = df3['sepal width (cm)']

plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.title("SVM_iris")

plt.scatter(x1points,y1points,label=m_label1)
plt.scatter(x2points,y2points,label=m_label2)
plt.scatter(x3points,y3points,label=m_label3)

plt.legend()
plt.show()
```

```python
# Train_test_data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# SVM_MODEL Import:

from sklearn.svm import SVC
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
model = SVC()
model.fit(x_train,y_train)
```

```python
# Testing Accuracy
model.score(x_train,y_train)
```

```python
'''
c - Regularization
c= 1,2,3,4,5,6 to n numbers,this is the parameter to set
and find out higher accuracy rate

'''
```

```python
# parameter check
svm1_model=SVC(C=10)
svm1_model.fit(x_train,y_train)
```

```python
# Training Accuracy Test Data
svm1_model.score(x_test,y_test)
```
